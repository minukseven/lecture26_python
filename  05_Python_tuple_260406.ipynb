### 튜플의 개념

```python
t1 = (1, 2, 3)
t2 = ('a', 'b', ('ab', 'cd'))
t3 = ('tuple', (1, 2), [3, 4])

print (t1)
```

```python
t4 = ()
t5 = (1,)

t6 = 1, 2, 3
t7 = 1,

print(t6)
x, y, z = t6
print(x, y, z)
```

### 튜플의 활용

```python
def add_and_mul(a, b):
    return a+b, a*b

result = add_and_mul(3, 5)
print(result)
```
