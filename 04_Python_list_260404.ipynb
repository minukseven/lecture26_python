### 리스트의 항목에 접근하기

```python
test_list = [1, 2, 3, 4, 5, 6, 7, 8,]

print(test_list[0:5])
print(test_list[:5])

print(test_list[1::3])
print(test_list[:5:2])
print(test_list[::3])
print(test_list[:])
print(test_list[::])

print(test_list[-1:-7:-2])
print(test_list[1:-1:])
print(test_list[-3::-1])
print(test_list[::-1])
```

### 리스트에 항목 추가하기

```python
a_list = ['one', 'two', 'three', 'for']
print('원본 데이터', a_list)

a_list.reverse()
print('역순 후 데이터', a_list)

a_list.sort()
print('정렬 후 데이터', a_list)

a_list = ['one', 'two', 'three', 'four']
b_list = sorted(a_list)
print('원본 데이터', a_list)
print('정렬 데이터', b_list)
```

```python
scores = [95, 45, 21, 74, 55]
print(sorted(scores, reverse=True))

print(scores)
scores.sort(reverse=True)
print(scores)
```

```python
# 1 ~ N의 수로 이루어진 리스트 생성
N = int(input("숫자 :"))
num_list = []
for i in range(1,N+1):
    num_list.append(i)
print(num_list)
```

```python
# 1 ~ N의 수로 이루어진 리스트 생성
N = int(input("숫자 :"))
num_list = [i for i in range(1,N+1)]
print(num_list)

# expression
num_list = [i * 2 for i in range(1,N+1)]
print(num_list)

# if문 사용
N = int(input("숫자 :"))
num_list = [i for i in range(1,N+1) if i % 2 ==0]
print(num_list)
```

### getIndex

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

def getIndex(num_list, target):
    for i in range(len(num_list)):
        if num_list[i] == target:
            return i
    return -1

print(getIndex(number_list, 25))
```

### getMax

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

def getMax(num_list):
    max_val = num_list[0]
    for num in num_list:
        if num > max_val:
            max_val = num
    return max_val

print(getMax(number_list))
```

### getMin

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

def getMin(num_list):
    min_val = num_list[0]
    for num in num_list:
        if num < min_val:
            min_val = num
    return min_val

print(getMin(number_list))
```

### countGT

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

def countGT(num_list, target):
    count = 0
    for num in num_list:
        if num > target:
            count += 1
    return count 

print(countGT(number_list, 42))
```

### sumList

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

def sumList(num_list):
    total = 0
    for num in num_list:
        total += num
    return total

print(sumList(number_list))
```

### SwapList

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

def swapList(num_list):
    n = len(num_list)
    for i in range(n // 2):
        num_list[i], num_list[n - 1 - i] = num_list[n - 1 - i], num_list[i]
```
